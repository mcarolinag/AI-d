```python
corpus='ProjectTitle: '+df['ProjectTitle']+ ' '+'ShortDescription: '+df['ShortDescription']+ ' '+'LongDescription: '+df['LongDescription']+' Province: '+df['Province']
```

```python
query_corpus='ProjectTitle: '+query['ProjectTitle']+ ' '+'ShortDescription: '+query['ShortDescription']+ ' '+'LongDescription: '+query['LongDescription']+' Province: '+query['Province']
```

```python


embedder = SentenceTransformer("intfloat/multilingual-e5-large-instruct")
```

Experiment 1: Concat all text columns

```python
# Use "convert_to_tensor=True" to keep the tensors on GPU (if available)
corpus_embeddings = embedder.encode(corpus.values, convert_to_tensor=True)
```

```python
torch.save(corpus_embeddings , "corpus_embeddings.pt")
```

```python
corpus_embeddings=torch.load('corpus_embeddings.pt')
```

```python
corpus_embeddings = embedder.encode(corpus.values, convert_to_tensor=True)
```

```python


# Find the closest 5 sentences of the corpus for each query sentence based on cosine similarity
top_k = min(5, len(corpus))
for query_ in query_corpus.values:
    query_embedding = embedder.encode(query_, convert_to_tensor=True)

    # We use cosine-similarity and torch.topk to find the highest 5 scores
    similarity_scores = embedder.similarity(query_embedding, corpus_embeddings)[0]
    scores, indices = torch.topk(similarity_scores, k=top_k)

    print("\nQuery:", query_)
    print("Top 5 most similar sentences in corpus:")

    for score, idx in zip(scores, indices):
        print(idx,corpus.values[idx], f"(Score: {score:.4f})")
```

Experiment 2 specifying max distance

```python
import geopy.distance
```

```python


# Find the closest 5 sentences of the corpus for each query sentence based on cosine similarity
top_k = min(5, len(corpus))
max_distance=200#km
for i in [0,1]:#range(query_corpus.values.shape[0]):

    query_embedding = embedder.encode(query_corpus.values[i], convert_to_tensor=True)
    coords_query = (query['Latitude'][i], query['Longitude'][i])


    df['distances']= df[['Latitude','Longitude']].apply(lambda x: geopy.distance.geodesic(coords_query, (x['Latitude'],x['Longitude'])).km, axis=1)

    max_dist_projects=df[df['distances']<=max_distance].reset_index()

    # We use cosine-similarity and torch.topk to find the highest 5 scores
    similarity_scores = embedder.similarity(query_embedding, corpus_embeddings[max_dist_projects['index'].values,:])[0]
    scores, indices = torch.topk(similarity_scores, k=top_k)

    print("\nQuery:", query_corpus.values[i])
    print("Top 5 most similar sentences in corpus:")

    for score, idx in zip(scores, max_dist_projects.iloc[indices.tolist()]['index'].values):
        print(idx,corpus.values[idx], f"(Score: {score:.4f})")
```
